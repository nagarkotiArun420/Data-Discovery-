```python
# Arrange the complaint types according to their average 'Request_Closing_Time', 
#categorized by various locations. Illustrate it through graph as well.    

#calculate average request closing time

avgClosingHours = dataCleanedUp_df.groupby(['Complaint Type' ,'Location Type'])['Request_Closing_Time'].mean().reset_index()
```

```python
# Convert datetime to hours

avgClosingHours['Request_Closing_Hours'] = avgClosingHours['Request_Closing_Time'].dt.total_seconds()/ 3600
```

```python
#pivot the table
pivotTable = avgClosingHours.pivot(index='Complaint Type',columns='Location Type', values="Request_Closing_Hours")

# Plot the columns
pivotTable.plot(kind='bar',figsize=(13, 8), width=1.75)
plt.xlabel('Time in hours')
plt.xticks(rotation=20, ha='right');
plt.tight_layout()
```

```python
from scipy.stats import f_oneway

# Null Hypothesis (H0): There is no difference in average repsonse time across the complain types
# Alternative Hypothesis (H1): There is a significant difference in average response time across the complaint types

# Created new dataframe to store Request_Closing_Hours in hours
dataCleanedUp_df.loc[:, 'Request_Closing_Hours'] =  dataCleanedUp_df['Request_Closing_Time'].dt.total_seconds() / 3600

#Group data by complaint type
groupedColumns = [group['Request_Closing_Hours'] for name, group in dataCleanedUp_df.groupby('Complaint Type')]

# Perform ANOVA test 
f_statistics, p_value = f_oneway(*groupedColumns)

print("f_statistics:",f_statistics)
print("p_value:", p_value)

#Significance lvel
alpha = 0.05

if p_value < alpha:
    print(f"\nAs the p_value {p_value} is less than the significance level {alpha}")
    print(f"There is difference in average response time across complaints, we reject the Null hypothesis.")

else:
    print(f"\nAs the p_value {p_value} is More than the significance level {alpha}")
    print(f"There is no difference in average response time across complaint types we fail to reject the Null hypothesis hypothesis.")
```

```python
from scipy.stats import chi2_contingency
```

```python
# Null Hypothesis (H0): The relationship between Complaint Type and Location Type is independent.
# Alternative Hypothesis (H1): The relationship between Complaint Type and Location Typ is dependent.

# Contingency table

contingency_table = pd.crosstab(dataCleanedUp_df['Complaint Type'], dataCleanedUp_df['Location Type'])

# perform the chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print results
print("Chi-Square stat:",chi2_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)

  # Significance level
alpha = 0.05

if p_value < alpha:
    print(f"\nAs we can see the p_value {p_value} is less than the signigicance level {alpha}")
    print(f"The relationship between Complaint Type and Location Type is dependent, We reject NUll hypothesis.")
else:
    print(f"\nAs we can see the p_value {p_value} is more than the signigicance level {alpha},")
    print(f"nThe relationship between Complaint Type and Location Type is Independent, We fail to reject NUll hypothesis.")
  
```
